```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
```

```python
dfx=pd.read_csv('Logistic_X_Train.csv')
dfy=pd.read_csv('Logistic_Y_Train.csv')
dfx_test=pd.read_csv('Logistic_X_Test.csv')

x=dfx.values
y=dfy.values
x_test=dfx_test.values
```

```python
x=(x-x.mean())/x.std()
x_test=(x_test-x_test.mean())/x_test.std()
```

```python
def hypothesis(x,w,b):
    hx=np.dot(x,w)+b
    return sigmoid(hx)
```

```python
def sigmoid(hx):
    return 1.0/(1.0+np.exp(-1.0*hx))
```

```python
def error(y,x,w,b):
    m=x.shape[0]
    err=0.0
    for i in range(m):
        hx=hypothesis(x[i],w,b)
        err+=y[i]*np.log2(hx)+(1-y[i])*np.log2(1-hx)
    return err/m
```

```python
def get_grad(x,w,b,y):
    grad_b=0.0
    grad_w=np.zeros(w.shape)
    m=x.shape[0]
    for i in range(m):
        hx=hypothesis(x[i],w,b)
        grad_w+=(y[i]-hx)*x[i]
        grad_b+=(y[i]-hx)
    grad_w/=m
    grad_b/=m
    return [grad_w,grad_b]
```

```python
def gradient_ascent(x,y,w,b,learning_rate=0.001):
    err=error(y,x,w,b)
    [grad_w,grad_b]=get_grad(x,w,b,y)
    w+=learning_rate*grad_w
    w+=learning_rate*grad_b
    return err,w,b
```

```python
def predict(x,w,b):
    confidence=hypothesis(x,w,b)
    if confidence<0.5:
        return 0
    else:
        return 1
```

```python
def get_acc(x_tst,y_tst,w,b):
    y_pred=[]
    for i in range(y_tst.shape[0]):
        p=predict(x_tst[i],w,b)
        y_pred.append(p)
    y_pred=np.array(y_pred)
    return float((y_pred==y_tst).sum())/y_tst.shape[0]
```

```python
loss=[]
acc=[]

w=2*np.random.random((x.shape[1],))
b=5*np.random.random()

for i in range(1000):
    l,w,b=gradient_ascent(x,y,w,b,learning_rate=0.01)
    acc.append(get_acc(x,y,w,b))
    loss.append(l)
```

```python
plt.plot(loss)
plt.ylabel('Negative of LofLikelihood')
plt.xlabel('Time')
plt.show()
```
